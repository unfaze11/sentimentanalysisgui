```python
# Importing required libraries
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tkinter import *
from tkinter import messagebox, filedialog
import tkinter.font as tkFont

#clear all fields and text area
def clearAll():
    negativeField.delete(0, END)
    neutralField.delete(0, END)
    positiveField.delete(0, END)
    overallField.delete(0, END)
    textArea.delete(1.0, END)

#detect and display sentiment
def detect_sentiment():
    sentence = textArea.get("1.0", "end").strip()  # Get the text and strip it
    if not sentence:
        messagebox.showwarning("Warning", "Please enter a sentence for sentiment analysis.")
        return
    
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)

    # Displaying sentiment results
    negativeField.delete(0, END)
    negativeField.insert(10, f"{sentiment_dict['neg']*100:.1f}% Negative")
    
    neutralField.delete(0, END)
    neutralField.insert(10, f"{sentiment_dict['neu']*100:.1f}% Neutral")

    positiveField.delete(0, END)
    positiveField.insert(10, f"{sentiment_dict['pos']*100:.1f}% Positive")

    overallField.delete(0, END)
    if sentiment_dict['compound'] >= 0.05:
        overallField.insert(10, "Positive")
    elif sentiment_dict['compound'] <= -0.05:
        overallField.insert(10, "Negative")
    else:
        overallField.insert(10, "Neutral")

# Function to open a file and load its content into the text area
def load_file():
    file_path = filedialog.askopenfilename(
        title="Open a Text File",
        filetypes=[("Text Files", "*.txt"), ("All Files", "*.*")]
    )
    if file_path:
        with open(file_path, "r") as file:
            text = file.read()
        textArea.delete(1.0, END)  
        textArea.insert(1.0, text)  

if __name__ == "__main__":
    gui = Tk()
    gui.title("Sentiment Detector")
    gui.config(background="light blue")

    font = tkFont.Font(family="Helvetica", size=12)

    input_frame = Frame(gui, bg="light blue")
    output_frame = Frame(gui, bg="light blue")

    Label(input_frame, text="Enter Your Sentence", bg="light blue", font=font).pack(pady=5)
    textArea = Text(input_frame, height=5, width=30, font="lucida 13")
    textArea.pack(pady=5)

    Button(input_frame, text="Load File", fg="Black", bg="Red", font=font, command=load_file).pack(pady=5)
    Button(input_frame, text="Check Sentiment", fg="Black", bg="Red", font=font, command=detect_sentiment).pack(pady=5)

    Label(output_frame, text="Negative:", bg="light blue", font=font).pack(pady=5)
    negativeField = Entry(output_frame, font=font)
    negativeField.pack(pady=5)

    Label(output_frame, text="Neutral:", bg="light blue", font=font).pack(pady=5)
    neutralField = Entry(output_frame, font=font)
    neutralField.pack(pady=5)

    Label(output_frame, text="Positive:", bg="light blue", font=font).pack(pady=5)
    positiveField = Entry(output_frame, font=font)
    positiveField.pack(pady=5)

    Label(output_frame, text="Overall Sentiment:", bg="light blue", font=font).pack(pady=5)
    overallField = Entry(output_frame, font=font)
    overallField.pack(pady=5)

    Button(output_frame, text="Clear", fg="Black", bg="Red", font=font, command=clearAll).pack(pady=5)
    Button(output_frame, text="Exit", fg="Black", bg="Red", font=font, command=gui.quit).pack(pady=5)

    input_frame.pack(pady=10)
    output_frame.pack(pady=10)

    gui.mainloop()
```
